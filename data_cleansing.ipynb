```python
import pandas as pd
from pathlib import Path
```

```python
jan_path = Path('Resources/JC-202201-citibike-tripdata.csv')
apr_path = Path('Resources/JC-202204-citibike-tripdata.csv')
jul_path = Path('Resources/JC-202207-citibike-tripdata.csv')
oct_path = Path('Resources/JC-202210-citibike-tripdata.csv')

jan_df = pd.read_csv(jan_path)
apr_df = pd.read_csv(apr_path)
jul_df = pd.read_csv(jul_path)
oct_df = pd.read_csv(oct_path)

jan_df.head()
```

```python
jan_df['season'] = "winter"
jan_df.head()
```

```python
apr_df['season'] = "spring"
apr_df.head()
```

```python
jul_df['season'] = "summer"
jul_df.head()
```

```python
oct_df['season'] = 'autumn'
oct_df.head()
```

```python
jan_apr_df = pd.concat([jan_df, apr_df], ignore_index=True)
jan_apr_jul_df = pd.concat([jan_apr_df, jul_df], ignore_index=True)
citibike_year_df = pd.concat([jan_apr_jul_df, oct_df], ignore_index=True)
citibike_year_df.head()
```

```python
citibike_year_df.tail()
```

```python
citibike_year_df.info()
```

```python
citibike_year_df.dropna()
citibike_year_df.info()
```

```python
citibike_year_df['started_at'] = pd.to_datetime(citibike_year_df['started_at'])
citibike_year_df['ended_at'] = pd.to_datetime(citibike_year_df['ended_at'])
citibike_year_df.info()
```

```python
citibike_year_df['ride_time'] = citibike_year_df['ended_at'] - citibike_year_df['started_at']
citibike_year_df.head()
```

```python
#code taken from https://stackoverflow.com/questions/27928/calculate-distance-between-two-latitude-longitude-points-haversine-formula
from math import cos, asin, sqrt, pi

def distance(row):
    lat1 = row['start_lat']
    lon1 = row['start_lng']
    lat2 = row['end_lat']
    lon2 = row['end_lng']
    r = 6371 # km
    p = pi / 180

    a = 0.5 - cos((lat2-lat1)*p)/2 + cos(lat1*p) * cos(lat2*p) * (1-cos((lon2-lon1)*p))/2
    return 2 * r * asin(sqrt(a))

citibike_year_df['ride_distance'] = citibike_year_df.apply(distance, axis=1)
citibike_year_df.tail()
```
